```python
#Converting price to a float
DFListFull['price'] = DFListFull['price'].str.replace('$', '')
DFListFull['price'] = DFListFull['price'].str.replace(',', '').astype('float64')
```

```python
#Based on the above correlations and a sense test, we select the follwing features for our models
DFListFull.head()
selected_columns = [
                'id',
                'host_response_rate',
                'host_is_superhost',
                'zipcode',
                'property_type',
                'room_type',
                'accommodates',
                'bathrooms',
                'bedrooms',
                'beds',
                'bed_type',
                'amenities',
                'price',
                'guests_included',
                'review_scores_rating',
                'review_scores_accuracy',
                'review_scores_cleanliness',
                'review_scores_checkin',
                'review_scores_communication',
                'review_scores_location',
                'review_scores_value'
]

DFListFull.drop( 
    [key for key in DFListFull.keys() if key not in selected_columns] ,
    axis=1, 
    inplace=True)
```

```python
#Since amenities is a textual feature, and we know longer the amenities means more the amenities which should be directly correlated to the price. So calculating the length of the amenities
DFListFull.loc[:,'amenities_length'] = DFListFull['amenities'].str.len()

```

```python
#Dropping the original amenities column
DFListFull = DFListFull.drop(['amenities'], 1)
```

```python
DFListFull
```

```python
#Converting the true/false values to binary 0/1
DFListFull['host_is_superhost'].replace('t',1, inplace=True)
DFListFull['host_is_superhost'].replace('f',0, inplace=True)
```

```python
DFListFull
```

```python
#Converting acceptance/response percentages to numbers
DFListFull['host_response_rate'] = DFListFull['host_response_rate'].replace('%','',regex=True).astype('float64')/100.00
```

```python
DFListFull
```

```python
#Recoding property_type values 
def recode(value):
    if value not in ['House', 'Apartment', 'Townhouse', 'Condominium']:
        return 'Other'
    return value

DFListFull['property_type'] = DFListFull['property_type'].apply(recode)
```

```python
DFListFull
```

```python
#Recoding bed_type values
def recode(value):
    if value not in ['Real Bed']:
        return 'Other'
    return value

DFListFull['bed_type'] = DFListFull['bed_type'].apply(recode)
```

```python
DFListFull
```

```python
#Filling the empty values in the columns
DFListFull['bathrooms'].fillna(method='ffill', inplace=True)
DFListFull['bedrooms'].fillna(method='ffill', inplace=True)
DFListFull['beds'].fillna(method='ffill', inplace=True)
DFListFull['host_is_superhost'].fillna(value=0,inplace=True)
DFListFull['host_response_rate'].fillna(value=0,inplace=True)
```

```python
#Creating distribution charts for all the numbered features
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numdf = DFListFull.select_dtypes(include=numerics)
numdf = numdf.dropna(axis=0, how='any')
numdf_variables = list(numdf)

graphs = len(numdf_variables)-1

position = []
for i in range(6):
    for j in range(3):
        b = i,j
        position.append(b)
        
fig, axes = plt.subplots(nrows=6, ncols=3, sharey=False, sharex=False, figsize=(12,20))
fig.subplots_adjust(hspace=.5)

for i in range(graphs):
    sns.distplot(numdf[numdf_variables[i]], ax=axes[position[i]], kde=False)
```
